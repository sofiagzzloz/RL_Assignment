```python
#IMPORTS
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import poisson
import sys
```

```python
#Problem Parameters
class jcp:
    @staticmethod
    def max_cars():
        return 20
    
    @staticmethod
    def γ():
        return 0.9
    
    @staticmethod
    def credit_reward():
        return 10
    
    @staticmethod
    def moving_reward():
        return -2
    
    @staticmethod
    def free_move_cars():
        return 1
    
    @staticmethod
    def parking_threshold():
        return 10
    
    @staticmethod
    def parking_cost():
        return -4
    
    @staticmethod
    def move_cost(action):
        if action > 0:
            return jcp.moving_reward() * max(action - jcp.free_move_cars(), 0)
        return jcp.moving_reward() * abs(action)
```

```python
class poisson_:
    
    def __init__(self, λ):
        self.λ = λ
        
        ε = 0.01
        
        # [α , β] is the range of n's for which the pmf value is above ε
        self.α = 0
        state = 1
        self.vals = {}
        summer = 0
        
        while(True):
            if state == 1:
                temp = poisson.pmf(self.α, self.λ) 
                if(temp <= ε):
                    self.α+=1
                else:
                    self.vals[self.α] = temp
                    summer += temp
                    self.β = self.α+1
                    state = 2
            elif state == 2:
                temp = poisson.pmf(self.β, self.λ)
                if(temp > ε):
                    self.vals[self.β] = temp
                    summer += temp
                    self.β+=1
                else:
                    break    
        
        # normalizing the pmf (probability mass function), values of n outside of [α, β] have pmf = 0
        
        added_val = (1-summer)/(self.β-self.α)
        for key in self.vals:
            self.vals[key] += added_val
        
            
    def f(self, n):
        try:
            Ret_value = self.vals[n]
        except(KeyError):
            Ret_value = 0
        finally:
            return Ret_value
```

```python
# Poisson Distribution - examples
ex = poisson_(3)
print(ex.λ)
print(ex.α,ex.β)
print(ex.vals)
```

```python
#A class holding the properties of a location together
class location:
    
    def __init__(self, req, ret):
        self.α = req                             #value of lambda for requests
        self.β = ret                             #value of lambda for returns
        self.poissonα = poisson_(self.α)
        self.poissonβ = poisson_(self.β)
```

```python
#Location initialisation

A = location(3,3)
B = location(4,2)
```

```python
#Initializing the value and policy matrices. Initial policy has zero value for all states.

value = np.zeros((jcp.max_cars()+1, jcp.max_cars()+1))
policy = value.copy().astype(int)
```

```python
def apply_action(state, action):
    return [max(min(state[0] - action, jcp.max_cars()),0) , max(min(state[1] + action, jcp.max_cars()),0)]
```

## Expected Reward

```python
def expected_reward(state, action):
    global value
    """
    state  : It's a pair of integers, # of cars at A and at B
    action : # of cars transferred from A to B,  -5 <= action <= 5 
    """
    
    ψ = 0 #reward
    new_state = apply_action(state, action)
    
    # adding reward for moving cars from one location to another (which is negative) 
    
    ψ = ψ + jcp.move_cost(action)
    
    if new_state[0] > jcp.parking_threshold():
        ψ += jcp.parking_cost()
    if new_state[1] > jcp.parking_threshold():
        ψ += jcp.parking_cost()
    
    #there are four discrete random variables which determine the probability distribution of the reward and next state
    
    for Aα in range(A.poissonα.α, A.poissonα.β):
        for Bα in range(B.poissonα.α, B.poissonα.β):
            for Aβ in range(A.poissonβ.α, A.poissonβ.β):
                for Bβ in range(B.poissonβ.α, B.poissonβ.β):
                    """
                    Aα : sample of cars requested at location A
                    Aβ : sample of cars returned at location A
                    Bα : sample of cars requested at location B
                    Bβ : sample of cars returned at location B
                    ζ  : probability of this event happening
                    """

                    # all four variables are independent of each other
                    ζ = A.poissonα.vals[Aα] * B.poissonα.vals[Bα] * A.poissonβ.vals[Aβ] * B.poissonβ.vals[Bβ]
                    
                    valid_requests_A = min(new_state[0], Aα)
                    valid_requests_B = min(new_state[1], Bα)
                    
                    rew = (valid_requests_A + valid_requests_B)*(jcp.credit_reward())
                    
                    #calculating the new state based on the values of the four random variables
                    new_s = [0,0]
                    new_s[0] = max(min(new_state[0] - valid_requests_A + Aβ, jcp.max_cars()),0)
                    new_s[1] = max(min(new_state[1] - valid_requests_B + Bβ, jcp.max_cars()),0)
                    
                    #Bellman's equation
                    ψ += ζ * (rew + jcp.γ() * value[new_s[0]][new_s[1]])
                    
    return ψ
```

## Policy Evaluation

```python
def policy_evaluation():
    
    global value
    
    # here policy_evaluation has a static variable ε whose values decreases over time
    ε = policy_evaluation.ε
    
    policy_evaluation.ε /= 10 
    
    while(1):
        δ = 0
        
        for i in range(value.shape[0]):
            for j in range(value.shape[1]):
                # value[i][j] denotes the value of the state [i,j]
                
                old_val = value[i][j]
                value[i][j] = expected_reward([i,j], policy[i][j])
                
                δ = max(δ, abs(value[i][j] - old_val))
                #print('.+\b', end = '')
                #sys.stdout.flush()
        print('\r'+'δ =',δ, end = '')
        #sys.stdout.flush()
    
        if δ < ε:
            break
```

```python
#initial value of ε
policy_evaluation.ε = 50
```

## Policy Improvement

```python
def policy_improvement():
    
    global policy
    
    policy_stable = True
    for i in range(value.shape[0]):
        for j in range(value.shape[1]):
            old_action = policy[i][j]
            
            max_act_val = None
            max_act = None
            
            τ12 = min(i,5)       # if I have say 3 cars at the first location, then I can atmost move 3 from 1 to 2
            τ21 = -min(j,5)      # if I have say 2 cars at the second location, then I can atmost move 2 from 2 to 1
            
            for act in range(τ21,τ12+1):
                σ = expected_reward([i,j], act)
                if max_act_val == None:
                    max_act_val = σ
                    max_act = act
                elif max_act_val < σ:
                    max_act_val = σ
                    max_act = act
                
            policy[i][j] = max_act
            
            if old_action!= policy[i][j]:
                policy_stable = False
    
    return policy_stable
```

```python
def save_policy():
    save_policy.counter += 1
    ax = sns.heatmap(policy, linewidth=0.5)
    ax.invert_yaxis()
    # plt.savefig('policy'+str(save_policy.counter)+'.svg')
    plt.show()
    plt.close()
    
def save_value():
    save_value.counter += 1
    ax = sns.heatmap(value, linewidth=0.5)
    ax.invert_yaxis()
    # plt.savefig('value'+ str(save_value.counter)+'.svg')
    plt.show()
    plt.close()
```

## Main Loop

```python
save_policy.counter = 0
save_value.counter = 0
```

```python
while(True):
    policy_evaluation()
    ρ = policy_improvement()
    save_value()
    save_policy()
    if ρ == True:
        break
```

# Reflection – Jack’s Car Rental (Version 1)

## How the Policy Changes

### 1. The Policy Gets “Edgier” Around 10 Cars

The parking penalty only applies when a location has **more than 10 cars**.

That means the reward function suddenly drops at 11 cars. It’s not gradual, it’s a step penalty.

Because of that:
- 10 cars → no penalty.
- 11 cars → immediate -$4.

So the optimal action can change sharply when you're near 10 cars.  
The policy becomes very sensitive around that boundary.

Compared to Version 0, the policy looks less smooth and more cautious near this threshold.

---

### 2. The Free A → B Move Changes Behavior

Normally, moving cars costs $2 per car.

But now:
- The first car from **A → B** costs **$0**.

This makes small A → B moves more attractive.

Especially if:
- You’re close to 10 cars at A,
- And moving one car avoids the parking penalty.

Since that first move is free, it often becomes an easy decision.  
So the system tends to redistribute cars from A to B more frequently.

---

### 3. Asymmetry in Moving Costs

In Version 0, costs were symmetric in both directions.

In Version 1:
- Moving **A → B** has one free car.
- Moving **B → A** does not.

So the cost structure becomes asymmetric.

Even if demand is similar at both locations, the policy may prefer A → B moves because they are slightly cheaper.

That directional bias didn’t exist in the original version.

---

### 4. Non-Linear Costs Make the Value Function Less Smooth

In Version 0:
- Costs and rewards were linear.
- The value function was smooth.
- Policy changes were gradual.

In Version 1:
- The parking penalty is a step cost.
- It appears suddenly at 11 cars.

This makes the value function more “bumpy” near 10 cars.

Because of that, the policy becomes more cautious.  
It tries to avoid states that trigger the penalty, even if those states might generate slightly more rental revenue.

---

## Version 0 vs Version 1

### Version 0
- Moving cars always costs $2 each.
- No parking penalty.
- Symmetric costs.
- Smooth and balanced policy.

### Version 1
- First A → B move is free.
- Parking penalty above 10 cars.
- Asymmetric costs.
- Policy becomes threshold-aware and more cautious.

---

## Overall Takeaway

What I found interesting is how small non-linear changes in the reward function significantly change the structure of the optimal policy.

Just adding:
- One free move in one direction, and  
- A small step penalty,

Was enough to make the policy asymmetric, more sensitive near thresholds, and less smooth overall.

It shows how important reward design is in reinforcement learning — even small modifications can strongly influence behavior.
